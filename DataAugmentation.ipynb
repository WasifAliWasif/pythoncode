```python
#!pip install imgaug
import matplotlib.pyplot as plt
import random
import imgaug.augmenters as iaa
 
```

```python
def flip_image(img):
    hflip= iaa.Fliplr(p=1.0)
    input_hf= hflip.augment_image(img)
    return input_hf


def rotate_image(img):
    a= random.randrange(-20,20,10)
    b= random.randrange(-20,20,10)
    rot1 = iaa.Affine(rotate=(a,b))
    input_rot1 = rot1.augment_image(img)
    return input_rot1    
    

def crop_image(img):
    crop1 = iaa.Crop(percent=(0, 0.3)) 
    input_crop1 = crop1.augment_image(img)
    return input_crop1


def gaussion_noise_image(img):
    a= random.randrange(0,30,10)
    b= random.randrange(0,30,10)
    noise=iaa.AdditiveGaussianNoise(0,50)
    input_noise=noise.augment_image(img)
    return input_noise

def shearY_image(img):
    a= random.randrange(-20,20,10)
    b= random.randrange(-20,20,10)
    shear = iaa.ShearY(shear=(10,20))
    input_shear=shear.augment_image(img)
    return input_shear  

def shearX_image(img):
    a= random.randrange(-20,20,10)
    b= random.randrange(-20,20,10)
    shear = iaa.ShearX(shear=(10,20))
    input_shear=shear.augment_image(img)
    return input_shear       

def elastic_image(img):
    elastic = iaa.ElasticTransformation(alpha=5.0, sigma=0.5)
    input_elastic= elastic.augment_image(img)
    return input_elastic  

```

```python
def perfor_random_augmentation(img):
    functions = [elastic_image, shearX_image, shearY_image,gaussion_noise_image, crop_image,rotate_image,flip_image]
    aug_img=random.choice(functions)(img)
    return aug_img
```
