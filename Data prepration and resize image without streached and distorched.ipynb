```python
import os
import shutil

# Source directory containing subdirectories with images
source_directory = '/content/drive/MyDrive/Protein Dataset/Covid'

# Destination directory where images will be copied with new names
destination_directory = '/content/drive/MyDrive/Three_Class_Protein/Covid'

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# Initialize counter for new names
count = 1

# Loop through each subdirectory in the source directory
for subdir in os.listdir(source_directory):
    subdir_path = os.path.join(source_directory, subdir)

    # Check if it's a directory
    if os.path.isdir(subdir_path):
        # Loop through each file in the subdirectory
        for filename in os.listdir(subdir_path):
            # Check if it's a file and has an image extension
            if os.path.isfile(os.path.join(subdir_path, filename)) and filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
                # Construct the new filename with a unique counter
                new_filename = f"image_{count}.jpg"  # You can adjust the extension if needed
                count += 1

                # Copy the image file to the destination directory with the new name
                shutil.copyfile(os.path.join(subdir_path, filename), os.path.join(destination_directory, new_filename))
                print(f"File '{filename}' copied successfully as '{new_filename}'")

print("All images copied and renamed successfully.")
```

```python
from PIL import Image, ImageOps

def resize_image(image_path, target_size=(224, 224)):
    # Open the image
    image = Image.open(image_path)

    # Resize the image while preserving the aspect ratio
    image.thumbnail(target_size)

    # Create a new blank image with the target size
    new_image = Image.new("RGB", target_size, (255, 255, 255))

    # Calculate the position to paste the resized image
    position = ((target_size[0] - image.size[0]) // 2, (target_size[1] - image.size[1]) // 2)

    # Paste the resized image onto the new image
    new_image.paste(image, position)

    return new_image

# Example usage
input_image_path = "/content/drive/MyDrive/Binary_Classification_Protein_Images_Dataset/Covid/image_1.jpg"
output_image = resize_image(input_image_path)
output_image.save("resized_image.jpg")
```

```python

import os

def count_images_in_folder(folder_path):
    image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp']  # Add more extensions if needed
    image_count = 0

    for file_name in os.listdir(folder_path):
        if os.path.isfile(os.path.join(folder_path, file_name)):
            if any(file_name.lower().endswith(ext) for ext in image_extensions):
                image_count += 1

    return image_count

if __name__ == "__main__":
    folder_path = input("/content/drive/MyDrive/Three_Class_Protein/Covid ")
    if os.path.exists(folder_path):
        num_images = count_images_in_folder(folder_path)
        print(f"Number of images in the folder: {num_images}")
    else:
        print("Folder path does not exist.")
```

```python
import os
from PIL import Image, ImageOps

def resize_image(image_path, target_size=(224, 224)):
    try:
        # Open the image
        image = Image.open(image_path)

        # Resize the image while preserving the aspect ratio
        image.thumbnail(target_size)

        # Create a new blank image with the target size
        new_image = Image.new("RGB", target_size, (255, 255, 255))

        # Calculate the position to paste the resized image
        position = ((target_size[0] - image.size[0]) // 2, (target_size[1] - image.size[1]) // 2)

        # Paste the resized image onto the new image
        new_image.paste(image, position)

        return new_image
    except Exception as e:
        print(f"Error resizing image '{image_path}': {e}")
        return None

def copy_directory(source_dir, destination_dir):
    try:
        # Traverse through each directory and file in the source directory
        for root, dirs, files in os.walk(source_dir):
            for file in files:
                # Get the full path of the file
                file_path = os.path.join(root, file)

                # Resize the image and get the new image object
                resized_image = resize_image(file_path)

                if resized_image:
                    # Construct the destination file path
                    destination_file_path = os.path.join(destination_dir, os.path.relpath(file_path, source_dir))

                    # Create the directory structure if it doesn't exist
                    os.makedirs(os.path.dirname(destination_file_path), exist_ok=True)

                    # Save the resized image to the destination
                    resized_image.save(destination_file_path)
                    print(f"Resized image '{file_path}' saved to '{destination_file_path}'")

        print(f"All files in directory '{source_dir}' resized and copied to '{destination_dir}'")
    except Exception as e:
        print(f"Error copying directory: {e}")

# Example usage
source_directory = "/content/drive/MyDrive/Three_Class_Protein"
destination_directory = "/content/drive/MyDrive/Resized_Three_Class_Protein"

copy_directory(source_directory, destination_directory)
```
